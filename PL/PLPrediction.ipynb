## Data Preprocessing

```python
merged_df = get_pl_player_df_for_season('2022-23')
merged_df.head()
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
merged_df["Start"] = le.fit_transform(merged_df["Start"])
```

```python
merged_df.sample(10)
```

```python
from DataSci.features import lag_features, rolling_cols
numerical_cols = merged_df.select_dtypes(include=np.number, exclude=None).columns.to_list()
print(numerical_cols)
col_subset = numerical_cols[2:]
essential_cols = ["Player", "Season", "Date", "Squad", "Opponent", "opponent_difficulty", "team_difficulty", "total_points"]

dfs = []
for idx, (player, player_df) in enumerate(merged_df.groupby('Player')):
    raw_df = player_df.fillna(0).sort_values('Date')
    lagged_df = lag_features(raw_df, [1,2], subset=col_subset)
    average_df = rolling_cols(raw_df, 5, np.average, subset=col_subset)
    df = pd.concat([lagged_df, average_df, player_df[essential_cols]], axis=1)
    # display(df)
    dfs.append(df.dropna())
```

```python
lagged_df = pd.concat(dfs)
display(lagged_df.columns)
lagged_df.sample(7)
```

## Model

```python
from sklearn.metrics import explained_variance_score, mean_squared_error, median_absolute_error, r2_score
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor
import lightgbm as lgb

def display_metrics(y_pred, y):
    return dict(
        mse=mean_squared_error(y, y_pred),
        R2=r2_score(y, y_pred),
        median_abs_err=median_absolute_error(y, y_pred),
        explained_variance=explained_variance_score(y, y_pred)
    )

TARGET = "total_points"
```

```python
cat_cols = ["Venue", "Squad", "Opponent"]
numerical_cols = merged_df.select_dtypes(include=np.number, exclude=None).columns.to_list()
numerical_cols.remove(TARGET)
feature_cols = numerical_cols + cat_cols
merged_df[cat_cols] = merged_df[cat_cols].astype('category')
df = merged_df[feature_cols + [TARGET]].dropna().copy()

le = LabelEncoder()
for col in cat_cols:
    df[col] = le.fit_transform(df[col])
```

```python
X, y = df[feature_cols], df[TARGET]
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.1, random_state=42, shuffle=True)
X_val, X_test, y_val, y_test = train_test_split(X_test, y_test, test_size=0.5, random_state=42)

X, y = df[numerical_cols], df[TARGET]
X_num_train, X_num_test, y_num_train, y_num_test = train_test_split(X, y, test_size=0.1, random_state=42, shuffle=True)
X_num_val, X_num_test, y_num_val, y_num_test = train_test_split(X_num_test, y_num_test, test_size=0.5, random_state=42)
```

```python
rfr_params = dict(
    max_depth=24,
    min_samples_leaf=8,
)
rfr = RandomForestRegressor(**rfr_params)
print("Numerical cols only (oos metrics):")
rfr.fit(X_num_train, y_num_train)
display(display_metrics(rfr.predict(X_num_val), y_num_val))

print("With some categorical cols (oos metrics):")
rfr.fit(X_train, y_train)
display(display_metrics(rfr.predict(X_val), y_val))
```

```python
df = merged_df.dropna().copy()

X, y = df[feature_cols], df[TARGET]
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.1, random_state=42, shuffle=True)
# X_val, X_test, y_val, y_test = train_test_split(X_test, y_test, test_size=0.5, random_state=42)
```

```python
### using lightgbm
train_data = lgb.Dataset(X_train, label=y_train, categorical_feature=cat_cols)
lgb_param = dict(
    boosting_type='gbdt',
    max_bin=25,
    num_leaves=25,
    objective='mse',
    metric='rmse',
    feature_fraction=0.9,
    learning_rate=0.1,
    device_type='gpu'
)
num_round = 125
bst = lgb.train(lgb_param, train_data, num_round)   # ,nfold=5, return_cvbooster=True
display(display_metrics(bst.predict(X_test), y_test))
```

```python
train_data = lgb.Dataset(X_train, label=y_train, categorical_feature=cat_cols)
cv_results = lgb.cv(lgb_param, train_data, num_boost_round=350, nfold=5)
print(cv_results.keys())
# Display results
print('Current parameters:\n', lgb_param)
print('\nBest num_boost_round:', len(cv_results['valid rmse-mean']))
print('Best CV score:', cv_results['valid rmse-mean'][-1])
```

### modelling for forecasting

```python
df = lagged_df.copy()
cat_cols = ["Squad", "Opponent"]
feature_cols = df.select_dtypes(include=np.number, exclude=None).columns.to_list() + cat_cols
display(feature_cols)
le = LabelEncoder()
for col in cat_cols:
    df[col] = le.fit_transform(df[col])

X, y = df[feature_cols], df[TARGET]
X.loc[:, cat_cols] = X.loc[:, cat_cols].astype('category')
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.1, random_state=42, shuffle=True)
X_val, X_test, y_val, y_test = train_test_split(X_test, y_test, test_size=0.5, random_state=42)
```

```python
rfr_params = dict(
    max_depth=24,
    min_samples_leaf=8,
)
rfr = RandomForestRegressor(**rfr_params)
rfr.fit(X_train, y_train)
display(display_metrics(rfr.predict(X_val), y_val))
```

```python
df = lagged_df.copy()
cat_cols = ["Squad", "Opponent"]
feature_cols = df.select_dtypes(include=np.number, exclude=None).columns.to_list() + cat_cols
feature_cols.remove(TARGET)
display(feature_cols)

X, y = df[feature_cols], df[TARGET]
X[cat_cols] = X[cat_cols].astype('category')
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.1, random_state=42, shuffle=True)
X_val, X_test, y_val, y_test = train_test_split(X_test, y_test, test_size=0.5, random_state=42)
```

```python
train_data = lgb.Dataset(X_train, label=y_train, categorical_feature=cat_cols)
lgb_param = dict(
    boosting_type='gbdt',
    # max_bin=100,
    # num_leaves=25,
    max_depth=15,
    objective='mse',
    metric='rmse',
    feature_fraction=0.9,
    learning_rate=0.1,
    device_type='gpu',
    num_iterations=500
)
cv_results = lgb.cv(lgb_param, train_data, nfold=5)
print(cv_results.keys())
# Display results
print('Current parameters:\n', lgb_param)
print('\nBest num_boost_round:', len(cv_results['valid rmse-mean']))
print('Best CV score:', cv_results['valid rmse-mean'][-1])
```

```python
num_round = len(cv_results['valid rmse-mean'])
train_data = lgb.Dataset(X_train, label=y_train, categorical_feature=cat_cols)
val_data = lgb.Dataset(X_val, label=y_val, categorical_feature=cat_cols)
bst = lgb.train(lgb_param, train_data, num_round, valid_sets=val_data, callbacks=[lgb.early_stopping(stopping_rounds=25)])
display(display_metrics(bst.predict(X_test), y_test))
```

```python
import xgboost as xgb
```

```python
dtrain = xgb.DMatrix(data=X_train, label=y_train, enable_categorical=True)
dval = xgb.DMatrix(data=X_val, label=y_val, enable_categorical=True)
dtest = xgb.DMatrix(data=X_test, label=y_test, enable_categorical=True)
params = {
    'objective': 'reg:squarederror',  # Use 'binary:logistic' for binary classification
    'learning_rate': 0.1,
    'max_depth': 20,
    'min_child_weight': 1,
    'gamma': 0,
    'subsample': 0.8,
    'colsample_bytree': 0.8,
    'seed': 42,
    'num_rounds': 200,
    'tree_method': 'hist',
    'device' : "cuda"
}

num_rounds = 250
model = xgb.train(params, dtrain, num_rounds,
    verbose_eval=20,
    early_stopping_rounds=25, 
    evals=[(dval, 'valid')],
)
display(display_metrics(model.predict(dtest), y_test))
```
